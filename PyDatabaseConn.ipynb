```python
db_name = 'dvdrental'
db_pass ='postgresql'
db_host ='localhost'
db_user = 'postgres'
db_port = '5432'
```

```python
#importing package

import psycopg2 as ps
```

```python
#making connection to datbase

conn = ps.connect(database=db_name,password=db_pass,host=db_host,port=db_port,user=db_user)
```

```python
#for making communication with with database we need curser
cur = conn.cursor()
```

```python
# during excuting query if you get error like this "current transaction is aborted, commands ignored until 
#                                                      end of transaction block?"
#then just use rollback
cur.execute("ROLLBACK")
```

```python
#we select the customer table from the dvdrental database and exccute the cursor
cur.execute('SELECT customer_id,store_id,first_name,email from customer');
rows = cur.fetchall()
for row in rows:
    print('customer_id= ',row[0])
    print('store_id= ',row[1])
    print('first_name= ',row[2])
    print('email = ',row[3]);
conn.commit()
#conn.close()
```

* The commit() method is used to confirm the changes made by the user to the database. Whenever any change is made to the database using update or any other statements, it is necessary to commit the changes. If we donot use the commit() method after making any changes to the database, the database will not not be updated and changes will not be reflected.

```python
#using COUNT
cur.execute("SELECT COUNT (customer_id) FROM customer");
rows=cur.fetchall()
for row in rows:
    print("customer_id = ",row[0])
conn.commit()
#conn.close()
```

```python
#using LIMIT
cur.execute("SELECT customer_id, store_id,first_name, email FROM customer LIMIT 2");
rows =cur.fetchall()
for row in rows:
    print("customer_id =  ",row[0])
    print("store_id = ",row[1])
    print("first_name = ",row[2])
    print("email = ",row[3])
```

```python
#Using order by

cur.execute("SELECT customer_id, store_id,first_name, email FROM customer ORDER BY first_name DESC");
rows =cur.fetchall()
for row in rows:
    print("customer_id =  ",row[0])
    print("store_id = ",row[1])
    print("first_name = ",row[2])
    print("email = ",row[3])
```

```python
#Using Between

cur.execute("SELECT customer_id, store_id,first_name, email FROM customer WHERE customer_id BETWEEN 1 AND 100");
rows =cur.fetchall()
for row in rows:
    print("customer_id =  ",row[0])
    print("store_id = ",row[1])
    print("first_name = ",row[2])
    print("email = ",row[3])
```

```python
#using IN

cur.execute('''SELECT customer_id, store_id,first_name, email FROM customer WHERE 
                       customer_id IN (1,7,35,55,45) ORDER BY first_name''');
rows =cur.fetchall()
for row in rows:
    print("customer_id =  ",row[0])
    print("store_id = ",row[1])
    print("first_name = ",row[2])
    print("email = ",row[3])
```

```python
#using LIKE

cur.execute("SELECT customer_id, store_id,first_name, email FROM customer WHERE first_name LIKE 'Ju%'  ");
rows =cur.fetchall()
for row in rows:
    print("customer_id =  ",row[0])
    print("store_id = ",row[1])
    print("first_name = ",row[2])
    print("email = ",row[3])
```

```python
#SELECT Payment table from Daatbase

cur.execute(" SELECT payment_id, customer_id,rental_id, amount FROM payment");
rows =cur.fetchall()
for row in rows:
    print("payment_id =  ",row[0])
    print("customer_id = ",row[1])
    print("rental_id = ",row[2])
    print("amount = ",row[3])
```

```python
#Using AVG

cur.execute(" SELECT AVG(amount) As avarage_amount FROM payment");
rows = cur.fetchall()
for row in rows:
    print("avarage_amount = ",row[0])
```

```python
# using Round function

cur.execute(" SELECT ROUND(AVG(amount),2) As avarage_amount FROM payment");
rows = cur.fetchall()
for row in rows:
    print("avarage_amount = ",row[0])
```

```python
#Using Sum

cur.execute(" SELECT SUM(amount) As total_amount FROM payment");
rows = cur.fetchall()
for row in rows:
    print("total_amount = ",row[0])
```

```python
#Using MIN

cur.execute(" SELECT MIN(amount) As minimum_amount FROM payment");
rows = cur.fetchall()
for row in rows:
    print("minimum_amount = ",row[0])
```

```python
#Using MAX

cur.execute(" SELECT MAX(amount) As maximum_amount FROM payment");
rows = cur.fetchall()
for row in rows:
    print("maximum_amount = ",row[0])
```

```python
# Using Group By

cur.execute("SELECT customer_id , SUM(amount) total_amount FROM payment GROUP BY customer_id");
rows = cur.fetchall()
for row in rows:
    print("customer_id = ",row[0])
    print("total_amount = ",row[1])
    print("=======================")
    
```

```python
#Establishing the connection
conn = ps.connect(
   database=db_name, user=db_user, password=db_pass, host=db_host, port= db_port
)
#Creating a cursor object using the cursor() method
cursor = conn.cursor()


#Creating table as per requirement
sql ='''CREATE TABLE EMPLOYEE(
   FIRST_NAME CHAR(20) NOT NULL,
   LAST_NAME CHAR(20),
   AGE INT,
   SEX CHAR(1),
   INCOME FLOAT
)'''
cursor.execute(sql)
print("Table created successfully........")
conn.commit()
#Closing the connection
conn.close()
```
